```python
import numpy as np
import sys
sys.path.append('/home/lucas/Documents/Log_Analysis')
import analog
import matplotlib.pyplot as plt
```

# 1st estimation

```python
folder = '/home/lucas/Documents/Log_Analysis/Battery'
log_file = analog.pathfromQGC(folder,index=119)
info = analog.logextract(log_file,'battery_status')


time = info['time_bs']
voltage = info['battery_voltage']
```

```python
true_mean = 3.85*4
```

```python
plt.figure()
plt.plot(time,voltage)
plt.axhline(true_mean,color='r')
plt.xlabel('time (s)')
plt.ylabel('voltage (V)')
plt.grid()
np.var(voltage)
```

```python
sensor_mean = np.mean(voltage*1e3)
sensor_var = np.var(voltage*1e3)
sensor_std = np.std(voltage*1e3)
print(f'Sensor bias is {(sensor_mean - true_mean):.0f} volts.')
print(f'Sensor variance is {sensor_var:.0f} mv^2.')
print(f'Sensor standard deviation is {sensor_std:.2f} mv.')
print(np.std(voltage))
```
