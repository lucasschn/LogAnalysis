```python
import sys
sys.path.append('/home/lucas/Documents/Log_Analysis')
import analog
import matplotlib.pyplot as plt
import numpy as np
```

```python
folder = '/home/lucas/Documents/Log_Analysis/Logs/Snow Orange avoidance testing'
log_file = analog.pathfromQGC(folder,index='221')
print(log_file)
info = analog.logextract(log_file,['distance_sensor','vehicle_attitude_setpoint'])
print(info.keys())
```

```python
time=info['time_ds']
dist=info['dist']
time_vas = info['time_vas']
pitch_sp = info['pitch_sp']
```

```python
dist = dist[time<500]
time = time[time<500]
pitch_sp = pitch_sp[time_vas<500]
time_vas = time_vas[time_vas<500]
```

```python
%matplotlib notebook

plt.figure()
plt.subplot(211)
plt.plot(time[time > 350],dist[time > 350])
plt.ylabel('distance sensor (m)')
plt.grid()
plt.subplot(212)
plt.plot(time_vas[time_vas > 360], pitch_sp[time_vas > 360])
plt.grid()
plt.xlabel('time (s)')
plt.ylabel('pitch rate (rad/s)')
```
